```python
from typing import Sequence
import pandas as pd
import numpy as np
sequences = {}
```

```python
df_loading = pd.read_csv('./drive/MyDrive/vtb24_hktn/loadings.csv', index_col='dateTime')
```

```python
df_loading = df_loading[df_loading['salePointName'] != float("nan")]
```

```python
del sequences[list(sequences)[274]]
```

```python
df_loading.groupby(['salePointName']).count()
```

```python
from numpy.core.multiarray import array
for office in df_loading['salePointName'].unique():
  sequences[office] = array(df_loading[df_loading['salePointName'] == office]['clients']).astype(int)
```

```python
from numpy import hstack
for office in sequences:
  sequences[office] = sequences[office].reshape((len(sequences[office]),1))
```

```python
dataset = hstack(list([sequences[i] for i in sequences]))
```

```python
df_loading.values
```

```python
def create_dataset(dataset, n_steps, n_out):
  X, y = list(), list()
  for i in range(len(dataset)):
    seq_end = i + n_steps
    if seq_end + n_out > len(dataset):
      break
    seq_x, seq_y=dataset[i:seq_end, 0], dataset[seq_end:seq_end + n_out,0]
    seq_x = seq_x.reshape((len(seq_x), 1))
    X.append(seq_x)
    y.append(seq_y)
  return array(X), array(y)
```

```python
X, y = create_dataset(dataset, 72, 72)
```

```python
X.shape
```

```python
verbose, epochs, batch_size = 0, 20, 16
n_timesteps, n_features, n_outputs = X.shape[1], X.shape[2], y.shape[1]
```

```python
X
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers import LSTM
from keras.layers import RepeatVector
from keras.layers import TimeDistributed
```

```python
model = Sequential()
model.add(LSTM(200, activation='relu', input_shape=(n_timesteps, n_features)))
model.add(RepeatVector(n_outputs))
model.add(LSTM(200, activation='relu', return_sequences=True))
model.add(TimeDistributed(Dense(100, activation='relu')))
model.add(TimeDistributed(Dense(1)))
model.compile(loss='mse', optimizer='adam')
# fit network
model.fit(X, y, epochs=epochs, batch_size=batch_size, verbose=verbose)
```

```python
import pickle
with open('./drive/MyDrive/vtb24_hktn/model.pkl', 'wb') as f:
  pickle.dump(model, f)
```

```python
def forecast(model, history, n_input):
	# flatten data
	data = array(history)
	# data = data.reshape((data.shape[0]*data.shape[1], data.shape[2]))
	# retrieve last observations for input data
	input_x = data[-n_input:, :]
	# reshape into [1, n_input, n]
	input_x = input_x.reshape((input_x.shape[1], input_x.shape[0], 1))
	# forecast the next week
	yhat = model.predict(input_x, verbose=0)
	# we only want the vector forecast
	yhat = yhat[0]
	return yhat
```
